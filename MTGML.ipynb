```python
import pandas as pd
creatures = pd.read_csv('c1.csv', index_col=[0])
noncreatures = pd.read_csv('nc1.csv',index_col=[0])
```

```python
def price_buckets(row):
    if row['prices'] >= 50:
        return 4
    elif row['prices'] >=20:
        return 3
    elif row['prices'] >=10:
        return 2
    elif row['prices'] >=5:
        return 1
    else:
        return 0
```

```python
creatures['price_buckets'] = creatures.apply(lambda row:price_buckets(row), axis=1)
```

```python
creatures['price_buckets'].value_counts()
```

```python
noncreatures['price_buckets'] = noncreatures.apply(lambda row:price_buckets(row), axis=1)
```

```python
noncreatures['price_buckets'].value_counts()
```

```python
brothers_creatures['price_buckets'] = brothers_creatures.apply(lambda row: price_buckets(row), axis=1)
brothers_noncreatures['price_buckets'] = brothers_noncreatures.apply(lambda row: price_buckets(row),axis=1)
```

```python
brothers_creatures.to_csv('brothers_creatures.csv')
brothers_noncreatures.to_csv('brothers_noncreatures.csv')
```
